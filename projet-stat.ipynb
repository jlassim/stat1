**Introduction :**

**Thème : Analyse des facteurs de survie des passagers du Titanic**

***Présentation du thème:***

Le naufrage du Titanic, survenu le 15 avril 1912 lors de son voyage inaugural de Southampton à New York, est l'un des événements les plus tragiques de l'histoire maritime. Ce désastre a coûté la vie à près de 1500 personnes sur les 2200 à bord. L'analyse des données relatives à cette catastrophe offre une opportunité unique de comprendre les circonstances entourant les décès et les survivants.

***Objectif de l'analyse :***

L'objectif principal de cette analyse est d'identifier et de comprendre les facteurs qui ont influencé les chances de survie des passagers du Titanic. Pour ce faire, nous examinerons diverses caractéristiques des passagers, telles que leur âge, leur sexe, leur classe de billet, ainsi que d'autres facteurs qui pourraient avoir contribué à leur survie.

**Base de données:**



*  **Source :** [Lien de télechargement du dataset](https://www.kaggle.com/datasets/brendan45774/test-file)
*   **Année :** Bien que l'événement se soit produit en 1912, l'ensemble de données Titanic est une recréation basée sur des informations historiques disponibles.

*   **Échantillon étudié :** L'échantillon étudié comprend des données sur les passagers du Titanic. Chaque ligne de l'ensemble de données représente un passager individuel, avec des attributs tels que le sexe, l'âge, la classe de billet, le tarif payé, le nombre de membres de la famille à bord, le port d'embarquement, etc.
*   **Taille de l'échantillon :** 418 lignes de 12 colonnes .


*   **Les variables :**

      1-PassengerId :Un identifiant unique attribué à chaque passager.

      2-Survived :Une variable binaire qui indique si le passager a survécu (1) ou non (0).

      3-Pclass :La classe dans laquelle le passager voyageait (1ère, 2ème ou 3ème classe).

      4-Name :Le nom du passager.

      5-Sex :Le sexe du passager.

      6-Age :L'âge du passager.

      7-SibSp :Le nombre de frères et sœurs (ou conjoints) à bord du Titanic.

      8-Parch :Le nombre de parents/enfants à bord du Titanic.

      9-Ticket :Le numéro de billet du passager.

      10-Fare :Le tarif payé par le passager.

      11-Cabin :Le numéro de cabine du passager.

      12-Embarked :Le port d'embarquement du passager (C = Cherbourg, Q = Queenstown, S = Southampton).
*   **Les variables d’intérêt :**

  1.   **Variable endogène (Y) :**

      La variable endogène serait Survived. Nous nous intéressons à savoir si un passager a survécu ou non.
  2.   **Variables potentiellement explicatives (X) :**

      **Pclass :** On pourrait supposer que la classe dans laquelle un passager voyageait pourrait avoir une influence sur ses chances de survie, avec peut-être une relation négative (les passagers voyageant en première classe pourraient avoir de meilleures chances de survie).

      **Sexe :**On pourrait supposer qu'il existe une relation entre le sexe et les chances de survie, peut-être avec une hypothèse selon laquelle les femmes auraient de meilleures chances de survie que les hommes.

      **Age :** On pourrait envisager une relation où les enfants pourraient avoir des taux de survie plus élevés que les adultes.

      **SibSp** et **Parch :**On pourrait supposer que les passagers voyageant en famille pourraient avoir des taux de survie différents de ceux voyageant seuls.

      **Fare :** On pourrait supposer que le tarif payé par le passager pourrait être lié à sa classe et, par conséquent, influencer ses chances de survie.
    

*   **Hypothèses sur les relations statistiques :**

      ** On pourrait supposer que la classe sociale (Pclass) aurait un effet significatif sur les chances de survie, avec une relation négative.

      **On pourrait supposer que le sexe (Sex) aurait un effet significatif sur les chances de survie, avec une relation positive pour les femmes.

      **On pourrait supposer qu'il existe une relation en U entre l'âge (Age) et les chances de survie, ce qui signifie que les enfants et les personnes âgées pourraient avoir des taux de survie plus élevés que les adultes en âge de travailler.

      **On pourrait supposer qu'il existe une relation positive entre le nombre de parents/enfants (Parch) ou de frères et sœurs (SibSp) à bord et les chances de survie, car les passagers voyageant en famille pourraient avoir un soutien mutuel.

      **On pourrait supposer qu'il existe une relation négative entre le tarif payé (Fare) et les chances de survie, étant donné que les passagers voyageant en première classe (payant des tarifs plus élevés) pourraient avoir de meilleures chances de survie.

**Corrélation:**

Pour déterminer si certaines variables X sont corrélées entre elles dans le dataset pn peut calculer la matrice de corrélation entre les variables quantitatives. La matrice de corrélation indiquera les relations linéaires entre chaque paire de variables quantitatives.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Sélectionner uniquement les variables quantitatives
quantitative_vars = ['Age', 'SibSp', 'Parch', 'Fare','Pclass']

# Calculer la matrice de corrélation
correlation_matrix = titanic_data[quantitative_vars].corr()

# Tracer la heatmap de la matrice de corrélation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de corrélation des variables quantitatives')
plt.show()
```

Nous observons Une corrélation négative de (-0,49) suggère qu'il existe une tendance où, en moyenne, à mesure que l'âge des passagers augmente, la classe dans laquelle ils voyagent a tendance à diminuer.Cette corrélation peut être expliquée par le fait que les passagers plus jeunes, comme les enfants et les jeunes adultes, peuvent être plus susceptibles de voyager avec leur famille et être inclus dans des billets de classe supérieure.

Nous voyons une corrélation négative (-0.58)de la tarif (fare) avec la classe (Pclass), ce qui est attendu. Les passagers voyageant en première classe ont généralement payé des tarifs plus élevés que ceux voyageant en deuxième ou troisième classe. Cette corrélation est conforme aux normes socio-économiques de l'époque, où les classes de billets étaient associées à différents niveaux de luxe et de confort.

Nombre de parents/enfants (Parch) :
Comme mentionné, il existe une corrélation positive (0.31) avec le nombre de frères et sœurs/conjoints (SibSp), ce qui suggère une relation similaire entre ces deux variables.

Une corrélation positive de (0,34) suggère que l'âge des passagers augmente, le tarif payé a également tendance à augmenter.
Cette corrélation peut être attribuée à plusieurs facteurs. Par exemple, les passagers plus âgés peuvent être plus enclins à voyager dans des classes de billets supérieures, qui ont des tarifs plus élevés, pour bénéficier de davantage de confort et de services à bord. Ils peuvent également être plus susceptibles d'opter pour des cabines individuelles ou des accommodations de meilleure qualité, ce qui se refléterait dans un tarif plus élevé.

```python
#Test du chi-deux (chi-square) :  test utilisé pour évaluer l'indépendance entre deux variables catégorielles.
from scipy.stats import chi2_contingency

# le sexe et la survie
sex_survived_cross = pd.crosstab(titanic_data['Sex'], titanic_data['Survived'])
chi2, p_value, dof, expected = chi2_contingency(sex_survived_cross)
print("Valeur de chi2 :", chi2)
print("p-value :", p_value)
#Test t de Student : utilisé pour comparer les moyennes de la variable Y entre deux groupes définis par la variable X.
#âge et la survie
from scipy.stats import ttest_ind
survived_age = titanic_data[titanic_data['Survived'] == 1]['Age'].dropna()
not_survived_age = titanic_data[titanic_data['Survived'] == 0]['Age'].dropna()
t_stat, p_value = ttest_ind(survived_age, not_survived_age)
print("Test t de Student : t-statistic =", t_stat, ", p-value =", p_value)
```
